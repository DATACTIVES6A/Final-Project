```python
# Visualisasi Data Categorical
plt.figure(figsize=(12, 8))
for i in range(len(cat)):
    plt.subplot(2, 2, i+1)
    ax = sns.countplot(data=data, x=cat[i], color='gray', orient='h')
    plt.xlabel(cat[i])
    plt.tight_layout()

    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

**KORELASI PEARSON**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca dataset e-commerce ke dalam DataFrame
data = pd.read_csv('Train-1.csv')

# Hitung matriks korelasi
correlation_matrix = data.corr()

# Gambar heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Korelasi antara Feature dan Label')
plt.show()

```

**KORELASI SPEARMEN**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca dataset e-commerce ke dalam DataFrame
data = pd.read_csv('Train-1.csv')

# Hitung matriks korelasi dengan metode Spearman
correlation_matrix = data.corr(method='spearman')

# Gambar heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Korelasi Spearman antara Feature Numerik')
plt.show()
```

**KORELASI KENDALL**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca dataset e-commerce ke dalam DataFrame
data = pd.read_csv('Train-1.csv')

# Hitung matriks korelasi Kendall
correlation_matrix = data.corr(method='kendall')  # Menggunakan method='kendall' untuk korelasi Kendall

# Gambar heatmap dengan warna biru
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Korelasi Kendall antara Feature dan Label')
plt.show()
```

> Selain EDA, lakukan juga beberapa analisis dan visualisasi untuk menemukan suatu business insight. Tuliskan minimal 3 insight, dan berdasarkan insight tersebut jelaskan rekomendasinya untuk bisnis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Train.csv')
```

## Insight 1

```python
plt.figure(figsize = (6,4))
sns.barplot(data=data, x = 'Customer_care_calls', y = 'Cost_of_the_Product')
plt.xlabel('Customer Care Calls', fontsize = 12)
plt.ylabel('Cost of the product', fontsize = 12)
plt.title('How does the Cost of the product influence the Customer care calls?', fontsize = 20)
plt.show()
```

Harga produk berpengaruh terhadap jumlah customer care calss,dapat dilihat pada plot diatas yang menggambarkan bahwa semakin besar harga produk, maka semakin besar juga customer care calls

Hal tersebut sejalan dengan plot sebelumnya, lebih banyak diskon diberikan karena pengiriman terlambat, ini bisa menjadi alasan mengapa lebih banyak customer care calls dilakukan dengan harga produk yang lebih tinggi.

## Insight 2

```python
plt.figure(figsize = (6,4))
sns.barplot(data = data, x = 'Customer_care_calls', y = 'Discount_offered', ci = False)
plt.title('How do the customer calls impact the discount offered?', fontsize = 20)
```

Dari plot diatas bisa dilihat bahwa diskon tertinggi diberikan kepada orang-orang dengan jumlah panggilan layanan pelanggan terendah

Ketika banyak panggilan dilakukan, kita dapat melihat bahwa lebih sedikit diskon yang ditawarkan. Hal ini mungkin dikarenakan harga product yang tinggi dan produk harus dikirim dalam waktu yang lebih lama

## Insight 3

##Relation between Product Importance and Discount Offered

```python
sns.boxplot(x = 'Product_importance', y = 'Discount_offered', data= data, hue = 'Product_importance')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

Dari plot diatas bisa dilihat bahwa produk produk yang high importance diberikan discount yang lebih tinggi
Rekomendasi  yang bisa kami berikan adalah Implementasi program loyalitas yang memberikan diskon eksklusif kepada pelanggan yang sering berbelanja produk Anda yang penting. Ini dapat mendorong retensi pelanggan.
